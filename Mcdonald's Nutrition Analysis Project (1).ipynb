```python
drinks_df = menu_df[(menu_df['Category'] == "Smoothies & Shakes")]
```

```python
snacks_df = menu_df[(menu_df['Category'] == "Snacks & Sides")]
```

```python
snacks_df
```

```python
N = 13
ind = np.arange(N) 
width = 0.75
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = snacks_df['Sugars']
y = snacks_df['Saturated Fat']
ax.bar(ind, x, width, color='purple')
ax.bar(ind, y, width,bottom=x, color='blue')
ax.set_yticks(np.arange(0, 10, 12))
plt.show()
```
